## Data Wrangling

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'C:/Users/Bala Krishna Baddi/An IBM Data Science Professional certificate/Data Analysis with python/auto.csv'
df = pd.read_csv(filename,header=None)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
```

```python
print(df.shape)
df.head(2)
```

```python
import numpy as np
```

```python
df.replace('?',np.nan,inplace=True)
df
```

```python
#missing Data
import seaborn as sns

missing_data = df.isnull()
```

```python
#Counting Missing Values

for column in missing_data.columns.values.tolist():
    print(column,'\n',missing_data[column].value_counts(),'\n')
```

### Missed values...

<li> normalized-losses - 41
<li>num-of-doors  - 2
<li>bore - 4
<li>stroke - 4
<li>horsepower - 2
<li>peak-rpm - 2
<li>price - 4

```python
#replacing the normallized lossed with the mean value.....

average=df['normalized-losses'].astype(float).mean(axis=0)
df['normalized-losses'] = df['normalized-losses'].astype(float).fillna(average)
```

```python
#Replacing num of doors with top frequency.....
df['num-of-doors'].replace(np.nan,4,inplace=True)
```

```python
#Replacing Bore nulls with average...
bore_avg = df.bore.astype(float).mean(axis=0)
df.bore = df.bore.astype(float).fillna(bore_avg)
```

```python
#Replacing the null strokes with average...
stroke_avg = df.stroke.astype(float).mean()
df.stroke = df.stroke.astype(float).fillna(stroke_avg)
```

```python
#Relacing Hourse power with Mean Value...
hp_avg = df.horsepower.astype(float)
df.horsepower = df.horsepower.astype(float).fillna(hp_avg)
```

```python
#Relacing Peak RPM with Mean Value...
hp_avg = df['peak-rpm'].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float).fillna(hp_avg)
```

```python
df =df.dropna(subset=['price'],axis=0)
df.price = df.price.astype('float')
df.reset_index(drop=True,inplace=True)
df.tail(1) # Now all the Values are Clear There are no Null values in the Data SET
```

```python
#Replacing Column names with underscore...

headers = list(df.columns)
headers_new =[]
for column in headers:
    h = column.replace('-','_')
    headers_new.append(h)
df.columns = headers_new
df.head(2)
```

```python
df['highway_mpL100km'] = 235/df[["highway_mpg"]]
df['city_L100km'] = 235/df[["city_mpg"]]
df.head()
```

 ### Data Normalization

```python
df.length =df.length/df.length.max()
df.height = df.height/df.height.max()
df.width = df.width/df.width.max()
```

# Binning

```python
#changing the Data types of the Hourse power and replacing the null values with mean
df.horsepower.fillna(df.horsepower.mean(),inplace=True)
df.horsepower= df.horsepower.astype(int,copy=())
```

```python
plt.hist(df['horsepower'])

#set Labels....

plt.xlabel('Hourse power')
plt.ylabel('count')
plt.title('horse power bins')
```

```python
bins = np.linspace(min(df.horsepower),max(df.horsepower),num=4)
bins
```

```python
group_names = ['low','medium','high']
df['Horse_power_binned'] = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','Horse_power_binned']].head()
```

```python
df['Horse_power_binned'].value_counts()
```

```python
plt.bar(group_names, df["Horse_power_binned"].value_counts())
plt.xlabel('Hourse power')
plt.ylabel('Count')
plt.title('Hourse power')
```

```python
dummy_variable_1 = pd.get_dummies(df.fuel_type)
dummy_variable_1.rename(columns={'diesel':'Fuel_Type_diesel','gas':'Fuel_type_GAS'},inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df,dummy_variable_1],axis=1)
df.drop('fuel_type',axis=1,inplace=True)
df.head()
```

```python
dummy_variable_2 = pd.get_dummies(df.aspiration)
dummy_variable_2.rename(columns={'std':'aspiration_Std','turbo':'aspiration_Turbo'},inplace=True)
dummy_variable_2
```

```python
df =pd.concat([df,dummy_variable_2],axis=1)
df
```

```python
df.drop('aspiration',axis=1,inplace=True)
df.head()
```

```python
#Sacing the File to the File system...
df.to_csv('Automobiles_cleaned')
print("Your Data is cleaned and stored in you computer")
```
